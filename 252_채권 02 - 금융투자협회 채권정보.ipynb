# XML 들여쓰기로 출력
xml.dom.minidom 의 toprettyxml() 함수를 활용하면 손쉽게 XML 들여쓰기 할수 있다.

````python
import xml.dom.minidom

def xml_pretty(str_xml):
    node = xml.dom.minidom.parseString(str_xml)
    return node.toprettyxml(indent='    ', newl='')

print(xml_pretty(r.text))
````

# XML to DataFrame

````python
# XML to DataFrame

import xml.etree.ElementTree as ET
import pandas as pd

tree = ET.XML(r.content) 
all_records = []

element = tree.findall('message/BISComDspDatListDTO/BISComDspDatDTO')
for i, child in enumerate(element):
    record = {}
    for i, subchild in enumerate(child):
        record[subchild.tag] = subchild.text
    all_records.append(record)

df = pd.DataFrame(all_records)
df
````

# 함수로 정리

````python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
from datetime import datetime

payload_tmpl = """<?xml version="1.0" encoding="utf-8"?>
<message>
  <proframeHeader>
    <pfmAppName>BIS-KOFIABOND</pfmAppName>
    <pfmSvcName>BISLastAskPrcROPSrchSO</pfmSvcName>
    <pfmFnName>listDay</pfmFnName>
  </proframeHeader>
  <systemHeader></systemHeader>
<BISComDspDatDTO><val1>{the_day}</val1></BISComDspDatDTO></message>
"""

def get_bond_interest(the_day=None):
    # the_day: "YYYYMMDD"
    if the_day is None:
        the_day = datetime.now().strftime("%Y%m%d")
    payload = payload_tmpl.format(the_day=the_day)
    r = requests.post('http://www.kofiabond.or.kr/proframeWeb/XMLSERVICES/', data=payload)

    # XML to DataFrame
    tree = ET.XML(r.content) 
    all_records = []
    element = tree.findall('message/BISComDspDatListDTO/BISComDspDatDTO')
    for i, child in enumerate(element):
        record = {}
        for i, subchild in enumerate(child):
            record[subchild.tag] = subchild.text
        all_records.append(record)

    df = pd.DataFrame(all_records)

    # XML 문서의 항목 명칭과 컬럼명 매칭
    col_maps = [
        ('val1', '종류명'), 
        ('val2', '잔존기간'), 
        ('val3', '오전'), 
        ('val4', '오후'), 
        ('val5', '전일대비'), 
        ('val6', '전일'), 
        ('val7', '연중최고'), 
        ('val8', '연중최저')
    ]

    # 사용할 컬럼만 추출
    df = df[col_list].copy()

    # 컬럼 이름 바꾸기
    for col in col_maps:
        df.rename(columns={col[0] : col[1]}, inplace=True)

    # 컬럼의 타입 바꾸기 (str to float)
    float_columns = ['연중최저', '오전', '오후', '전일대비', '연중최고', '연중최저']
    df[float_columns] = df[float_columns].astype(float)
    
    # 인덱스 지정
    df.set_index('종류명', inplace=True)
    return df
````
