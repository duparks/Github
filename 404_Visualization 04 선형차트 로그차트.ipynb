```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from pandas_datareader import data, wb
from datetime import datetime
import requests
import json

plt.rcParams['figure.figsize'] = (16, 12)
```

## 1998년~2015년 코스피 지수

```python
code = '^KS11'  # 코스피 지수

start = datetime(1998, 1, 1)
end = datetime(2015, 12, 1)
df = data.get_data_yahoo(code, start, end)
df.head()
```

```python
ax = df['Adj Close'].plot(figsize=(18,6))
ax.grid(True)
```

## 선형 차트, 로그 차트

```python
ax = plt.subplot(2, 1, 1)
ax.plot(df.index, df['Adj Close'], 'b')

ax = plt.subplot(2, 1, 2)
ax.plot(df.index, df['Adj Close'], 'r')
ax.set_yscale('log')
ax.grid(True)
```

## tick 과 tick label 추가

```python
min, max = df['Adj Close'].min(), df['Adj Close'].max()
yticks = np.arange(min, max, 200)
yticks = yticks.astype(np.int)

ax = plt.subplot(2, 1, 1)
ax.plot(df.index, df['Adj Close'], 'b')
ax.set_ylim(min, max)
ax.set_yticks(yticks) 
ax.set_yticklabels(yticks)
ax.grid(True)

ax = plt.subplot(2, 1, 2)
ax.plot(df.index, df['Adj Close'], 'r')
ax.set_yscale('log')
ax.set_ylim(min, max)
ax.set_yticks(yticks) 
ax.set_yticklabels(yticks)
ax.grid(True)
```
