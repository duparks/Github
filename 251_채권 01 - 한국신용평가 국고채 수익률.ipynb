# 페이지 읽어오기
pd.read_html() 로 테이블을 모두 읽어온다

```python
import pandas as pd

url = 'https://www.kisrating.com/ratingsStatistics/statics_spread.do'
dfs = pd.read_html(url)
```

# 국고채 금리

```python
df_interest = dfs[0]
df_interest.set_index('구분', inplace=True)
df_interest
```

# 금리 스프레드


```python
df_spread = dfs[1]
df_spread.set_index('구분', inplace=True)
df_spread
```

# 특정 날짜의 국고채 금리

날짜를 지정하고 검색
* https://www.kisrating.com/ratingsStatistics/statics_spread.do
* startDt:2017.09.01
<img src="https://i.imgur.com/JiFo2Lt.png" >



```python
import pandas as pd
import requests

url = 'https://www.kisrating.com/ratingsStatistics/statics_spread.do'
data = {'startDt':'2016.09.01'}
r = requests.post(url, data=data)

dfs = pd.read_html(r.text)
df_interest = dfs[0]
df_interest.set_index('구분', inplace=True)
df_interest
```

```python
df_interest.loc['BBB+', '6월']
```
