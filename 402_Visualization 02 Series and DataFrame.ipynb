# Pandas Series.plot()

```python
%matplotlib inline

import pandas as pd
import numpy as np

s = pd.Series(np.random.randn(10).cumsum(0), index=np.arange(0, 100, 10))
s.plot()
```

```python
s.plot(ylim=(-8, 8))
```

## Series.plot()   옵션
* Series.plot() 옵션으로 ylim, ytics, xlim, xticks 등 활용

| 옵션 | 설명 | 예제 |
| ----- | ----- | ------ |
| label | 이름 | |
|  ax | 서브플롯 객체(다양한 설정을 ax에 지정해 두고 사용할때 쓴다) | |
|  style | 스타일 문자열. 예 'r--' | |
|  alpha | 투명도 (0~1사이 실수값) | |
|  kind | 그래프 종류 | 'line', 'bar', 'barh', 'kde' |
|  logy | y축 로그 스케일링 | |
|  use_index | 객체의 인덱스를 눈금의 이름으로 사용할지 여부 | |
|  rot | 눈금 회전 각도(0~360) | |
|  xticks | x축으로 사용할 값 | |
|  yticks | y축으로 사용할 값 | |
|  xlim | x축의 범위 | |
|  ylim | y축의 범위 | |
|  grid | 그리드 그리기 여부(기본값은 True) | |

```python
s = pd.Series(np.random.randn(10).cumsum(0))
s.plot()
```

## DataFrame

```python
df = pd.DataFrame(np.random.randn(10,2).cumsum(0),  
                  columns=['Rand-A', 'Rand-B'],  
                  index=np.arange(0, 100, 10))
df.plot()
```

## DataFrame.plot() 옵션

| 옵션  |  설명 | 예제 |
| ------- | ------------------------------------| --- |
| subplots | 각 컬럼을 하나의 서브플롯으로 그린다 |  |
| sharex | subplots=True이면 x축을 공유 |  |
| sharey | subplots=True이면 y축을 공유 |  |
| figsize | 크기를 튜플로 지정 |  |
| title | (문자열) 제목 |  |
| legend |(True)범례추가 여부 |  |
| sort_columns | 컬럼을 알파벳 순으로 그린다 |  |
